```python
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
color_images_path = 'Images/color'
grayscale_images_path = 'Images/gray'
```

```python
color_images_files = os.listdir(color_images_path)
grayscale_images_files = os.listdir(grayscale_images_path)

num_images_to_display = 5 

plt.figure(figsize=(10, 4))

for i in range(num_images_to_display):
    img_color = Image.open(os.path.join(color_images_path, color_images_files[i]))
    plt.subplot(2, num_images_to_display, i + 1)
    plt.imshow(img_color)
    plt.axis('off') 

    img_gray = Image.open(os.path.join(grayscale_images_path, grayscale_images_files[i]))
    plt.subplot(2, num_images_to_display, num_images_to_display + i + 1)
    plt.imshow(img_gray, cmap='gray')
    plt.axis('off')  

plt.tight_layout()
plt.show()
```

```python
print(f"Amount of color images: {len(color_images_files)}")
print(f"Amount of grayscale images: {len(grayscale_images_files)}")
```

```python
color_image = Image.open(os.path.join(color_images_path, color_images_files[0]))
width, height = color_image.size

print(f'Color image size: {width} x {height} pixels')
```

```python
grayscale_image = Image.open(os.path.join(color_images_path, grayscale_images_files[0]))
width_grayscale, height_grayscale = grayscale_image.size

print(f'Grayscale image size: {width_grayscale} x {height_grayscale} pixels')
```
